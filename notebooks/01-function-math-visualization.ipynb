# Problem

以 NumPy 產生函數與參數曲線的座標，再用 Matplotlib 將正弦函數與心形方程式視覺化。


# Method

- 使用 `np.linspace` 建立連續取樣點。
- 計算正弦函數與心形參數方程式的座標。
- 以 Matplotlib 繪製座標軸、曲線、填色與手工字樣。
- 本 Notebook 沒有模型訓練流程，因此不需要隨機種子。


```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. 畫函數的圖形

畫一個函數

$$f(x)=\sin(x)$$

```python
x = np.linspace(-10, 10, 400) # 從 -10 到 10 生成 400 個點
y = np.sin(x) # 計算每個 x 值的 sin 值
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = sin(x)', color='brown')
plt.title('Graph of f(x) = sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

## 3. 特別企劃：情人節心形曲線繪製

這部分我們將利用數學方程式繪製一個愛心圖形。這不僅是一個有趣的數學應用，也非常適合在特別的節日分享。

### 使用套件介紹

1. **NumPy**：
   - **功能**：提供強大的多維陣列運算功能。
   - **應用**：我們使用 `np.linspace` 產生 0 到 $2\pi$ 之間的連續數值，代表參數方程式中的 $\theta$。

2. **Matplotlib**：
   - **功能**：Python 最基礎且強大的資料視覺化工具。
   - **應用**：我們利用 `plt.fill` 來為心形塗上顏色，並使用 `plt.axis('equal')` 確保比例正確，不會讓愛心變形。

### 數學原理：心形參數方程式
我們使用的公式如下：
$$
x = 16 \sin^3(t)
$$
$$
y = 13 \cos(t) - 5 \cos(2t) - 2 \cos(3t) - \cos(4t)
$$
其中 $t$ 的範圍是 $0$ 到 $2\pi$。

```python
import numpy as np
import matplotlib.pyplot as plt

# 定義參數 t，生成 1000 個點
t = np.linspace(0, 2 * np.pi, 1000)

# 心形參數方程
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# 縮放數值以配合座標軸範圍
x_scaled = x * 0.5
y_scaled = y * 0.5

# 繪圖設定
plt.figure(figsize=(10, 10))
plt.fill(x_scaled, y_scaled, color='lightpink', edgecolor='deeppink', linewidth=2, alpha=0.5)

# --- 使用座標點連線生成文字 ---
def plot_letter(ax, ay, color='red'):
    plt.plot(ax, ay, color=color, linewidth=3)

t_o = np.linspace(0, 2*np.pi, 100)

# --- I (置中上方) ---
plot_letter([0, 0], [6, 8], color='red')
plot_letter([-0.5, 0.5], [8, 8], color='red')
plot_letter([-0.5, 0.5], [6, 6], color='red')

# --- LOVE (移至上方置中) ---
offset_x_love = -4
offset_y_love = 1
# L
plot_letter(offset_x_love + np.array([0, 0, 1.5]), offset_y_love + np.array([2, 0, 0]), color='red')
# O
plot_letter(offset_x_love + 3 + 0.7*np.cos(t_o), offset_y_love + 1 + 1*np.sin(t_o), color='red')
# V
plot_letter(offset_x_love + np.array([4.5, 5.5, 6.5]), offset_y_love + np.array([2, 0, 2]), color='red')
# E
plot_letter(offset_x_love + np.array([8.5, 7.5, 7.5, 8.5]), offset_y_love + np.array([2, 2, 0, 0]), color='red')
plot_letter(offset_x_love + np.array([7.5, 8.2]), offset_y_love + np.array([1, 1]), color='red')

# --- YOU (移至下方置中) ---
offset_x_you = -2.5
offset_y_you = -3
# Y
plot_letter(offset_x_you + np.array([-1.5, 0, 1.5]), offset_y_you + np.array([3, 1, 3]), color='darkred')
plot_letter(offset_x_you + np.array([0, 0]), offset_y_you + np.array([1, -1]), color='darkred')
# O
plot_letter(offset_x_you + 2.5 + 0.8*np.cos(t_o), offset_y_you + 1 + 1.2*np.sin(t_o), color='darkred')
# U
plot_letter(offset_x_you + np.array([4, 4, 6, 6]), offset_y_you + np.array([3, 1, 1, 3]), color='darkred')

# 設定範圍與美化
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='black', linewidth=0.5, alpha=0.3)
plt.axvline(0, color='black', linewidth=0.5, alpha=0.3)
plt.grid(True, linestyle='--', alpha=0.3)
plt.title('Mathematical Heart: I LOVE YOU', fontsize=16, color='deeppink')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```
